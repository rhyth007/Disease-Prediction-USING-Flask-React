```python
df = pd.read_csv('Training.csv')
```

```python
df.head()
```

```python
from  sklearn.preprocessing import  StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('prognosis',axis=1))
```

```python
scaled_features = scaler.transform(df.drop('prognosis',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test, y_train,y_test = train_test_split(scaled_features,df['prognosis'],test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
pickle.dump(knn,open("KNNmodel.pkl","wb"))
```
